```python
# ==========================================

# LOGIC BUILDING QUESTION - 1: FINISHED SUITE/tasksTASK1 inputINPUTThe first line of input is a single integer all the lines that follow contain integers each separated by space.

# ==========================================

# ------------------------------------------

# TASK 1: User Login Check

# ------------------------------------------

print("--- Task 1: Login Check ---")

username = "admin"

password = "1234"

if username == "admin" and password == "1234":

print("Login Successful")

else:

print("Invalid Credentials")

print("\n")

# ------------------------------------------

# TASK 2: Pass/Fail Analyser

# ------------------------------------------

print("--- Task 2: Pass/Fail Analyzer ---")

marks = [45, 78, 90, 33, 60]

pass_count = 0

fail_count = 0

for mark in marks:

if mark >= 50:

pass_count += 1

else:

fail_count += 1

print(f"Marks: {marks}")

print(f"Total Passed: {pass_count}")

print(f"Total Failed: {fail_count}")

print("\n")

# ------------------------------------------

# TASK 3: Simple Data Cleaner

# ------------------------------------------

print("Task 3: Data Cleaner")

names = [" Alice ", "bob", " CHARLIE "]

# List comprehension for resue cleaning causes speed up!

cleaned_names = [name. strip(). lower() for name in names]

print(f"Original: {names}")

print(f"Cleaned: {cleaned_names}")

print("\n")

# ------------------------------------------

# TASK 4: Length of Messaging Analysis

# ------------------------------------------

print("--- Task 4: Message Length Analyser ---")

messages = [“Hi”, “Welcome to the platform!”, “OK”]

for msg in messages:

length = len(msg)

status = "FLAGGED" if length > 10 else "OK"

print(f"Message: '{msg}' | Length: {length} -{'->' if status else ' Unknown '}-")

print("\n")

# ------------------------------------------

# Task 5: Error Message Detection

# ------------------------------------------

print("--- Challenge 5: Error Log Monitor ---")

logs=["INFO", "ERROR","WARNING","ERROR"]

# Using built-in 'count' method for optimization and cleanliness.

total_errors = logs. count("ERROR")

print(f"Logs: {logs}")

print(f"Total 'ERROR' counts: {total_errors}")

print("\n")

print("=== ALL TASKS SUCCEEDED ===")
```
