Problem 1: Social Media - Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like

    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

# Example Usage
analyze_engagement([450, 200, 400])
```

Problem 2: Healthcare - Medicine Stock Alert System

```python
def check_stock(medicine_name, current_stock):
    status = "Low Stock Alert" if current_stock < 10 else "Stock Sufficient"

    print(f"Medicine: {medicine_name}")
    print(f"Medicine Stock: {current_stock}")
    print(f"Status: {status}")

# Example Usage
check_stock("Paracetamol", 6)
```

Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def check_rainfall(rainfall_data, required_level):
    total = 0
    for val in rainfall_data:
        total += val

    avg_rainfall = total / len(rainfall_data)
    status = "Adequate Rainfall" if avg_rainfall >= required_level else "Inadequate Rainfall"

    print(f"Average Rainfall: {avg_rainfall:.1f}")
    print(f"Rainfall Status: {status}")

# Example Usage
check_rainfall([80, 70, 66], 70)
```

Problem 4: Social Media - Duplicate Account Detection

```python
def detect_duplicates(usernames):
    # Sets only store unique values; if lengths differ, duplicates exist
    has_duplicates = len(usernames) != len(set(usernames))
    result = "Yes" if has_duplicates else "No"

    print(f"Duplicate Accounts Found: {result}")

# Example Usage
detect_duplicates(["user123", "dev_mind", "user123"])
```

Problem 6: Agriculture – Premium Crop Price Filter


```python
def filter_premium_crops(price_list):
    # Using list comprehension for professional filtering
    premium_crops = [price for price in price_list if price > 2000]

    print(f"Premium Crops: {premium_crops}")

# Example Usage
filter_premium_crops([1500, 2500, 1800, 3200])
```

Problem 7: System Monitoring – Application Health Checker


```python
def monitor_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

# Example Usage
monitor_health(7)
```

Problem 9: E-Learning – Student Attendance Eligibility System


```python
def check_exam_eligibility(attendance_list):
    # attendance_list is a list of booleans or 1/0 (1 for present, 0 for absent)
    attendance_pct = (sum(attendance_list) / len(attendance_list)) * 100
    status = "Eligible" if attendance_pct >= 75 else "Not Eligible"

    print(f"Attendance Percentage: {attendance_pct:.1f}")
    print(f"Exam Eligibility: {status}")

# Example Usage (12 days present out of 15)
check_exam_eligibility([1, 1, 1, 0, 1, 1, 1, 1, 1, 0, 1, 1, 1, 0, 1])
```

Problem 10: Smart Electricity Bill Analyzer


```python
def calculate_electricity_bill(units):
    bill = 0

    # Logic for Slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Classification logic
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print(f"Total Bill: ₹{bill}")
    print(f"Usage Classification: {usage}")
    return bill, usage

# Example Usage
calculate_electricity_bill(250)
```
